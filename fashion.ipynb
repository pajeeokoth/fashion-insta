# Backlog

**Story 1** — Photo Upload & Wardrobe Capture: As a user, I want to upload photos of my clothing and outfits so the app can build a personal wardrobe catalog.

*Acceptance Criteria:* multiple photos accepted (camera/gallery); auto-tagging of garment type (top, bottom, outerwear, accessory) or manual edit; photos stored to user profile; privacy consent prompt shown.

*Priority:* High.

*Estimate:* 5 SP.

**Story 2** — Outfit Recommendation Engine: As a user, I want the app to suggest outfit combinations from my wardrobe for a given occasion and weather so I can choose what to wear quickly.

*Acceptance Criteria:* recommendations respect occasion (casual, work, formal), weather input or forecast integration, show 3–5 ranked outfits with confidence scores, each recommendation lists constituent items and styling notes.

*Priority:* High.

*Estimate:* 8 SP.

**Story 3** — Visual Similarity & Item Matching: As a user, I want the app to find visually similar items or alternative suggestions so I can replace missing or unliked pieces.

*Acceptance Criteria:* given a selected item, show visually similar items from my wardrobe or recommended catalog; similarity includes color, pattern, and silhouette; options to filter by color/season.

*Priority:* Medium.

*Estimate:* 5 SP.

**Story 4** — Virtual Try-On / Outfit Preview: As a user, I want to virtually preview outfits on my uploaded photo or avatar so I can see how combinations look before wearing them.

*Acceptance Criteria:* overlay outfits on user photo or avatar with reasonable alignment; toggle individual items on/off; save preview as favorite; fallback to flat-mock preview if advanced alignment unavailable.

*Priority:* Medium.

*Estimate:* 8–13 SP (depending on tech: simple overlay vs AR).

**Story 5** — Personalization & Learning: As a user, I want the app to learn my style preferences over time so recommendations improve with my feedback.

*Acceptance Criteria:* capture likes/dislikes and skip actions; recommendation model updates weights per user; provide a settings page to set style preferences (colors, formality, comfort); show a brief explanation why each suggestion was chosen.

*Priority:* High.

*Estimate:* 8 SP.

**Story 6** — Outfit Planner & Calendar Integration: As a user, I want to plan outfits for future dates/events and sync with my calendar so I can prepare ahead.

*Acceptance Criteria:* create outfit-for-date entries, view calendar of planned outfits, optionally sync to device calendar or export; conflict warnings for duplicate outfit use.

*Priority:* Medium.

*Estimate:* 5 SP.

**Story 7** — Social Sharing & Feedback: As a user, I want to share outfit options with friends and receive feedback so I can pick outfits with others’ opinions.

*Acceptance Criteria: *shareable links or in-app polls, friends can upvote/comment, privacy controls for shared content, store feedback linked to outfit for future personalization.

*Priority:* Low/Optional for MVP.

*Estimate:* 3–5 SP.

**Story 8** — Recommendations Diagnostics & Explainability: As a user, I want short explanations for recommendations so I understand why an outfit was suggested.

*Acceptance Criteria:* each recommendation includes 1–2 reasons (e.g., "color match", "weather-appropriate", "formal event"), and an option to provide feedback that updates preferences.

*Priority:* Medium.

*Estimate:* 3 SP.
**Story 9** — Multi-Device Sync & Cloud Storage: As a user, I want my wardrobe and preferences to sync across devices so I can access my data anywhere.
*Acceptance Criteria:* cloud storage of user data, real-time sync across devices, offline access with local caching, secure authentication.
*Priority:* High.
*Estimate:* 5 SP.
